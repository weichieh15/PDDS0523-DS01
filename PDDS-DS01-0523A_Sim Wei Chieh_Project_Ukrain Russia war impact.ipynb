```python
# We have already read the Dataset1 of Russia Ukraine war.
# Here you have to use import seaborn as sns library and sns.lineplot method.
# View the content of the dataset1.csv file, it is containing 285 entries and 18 columns.

# Import required packages: Matplotlib, Seaborn
import matplotlib.pyplot as plt
import seaborn as sns

# Read "Dataset1.csv", store them in "Act4Dataset1"
Act4Dataset1 = pd.read_csv('Dataset1.csv')

# Display the content of the dataset1.csv file, it is containing 285 entries and 18 columns.
print('Display the content of Act4Dataset1:')
Act4Dataset1
```

```python
# Display Act4Dataset1 information using info()
Act4Dataset1.info()
```

```python
# Display the list of columns associated with the dataset1 by using .columns property.
print('Display the list of columns associated with the dataset1 by using .columns property:\n\n', Act4Dataset1.columns)
```

```python
# Use plot() and show() method to visualize and identify the pattern of each attribute from the dataset.

# set figure size with 10 width by 10 height in inches
plt.figure(figsize=(10,10))
# Display each attribute from the dataset in one plot.
for col in Act4Dataset1.columns:
    # Only if attribute 'date', 'day', and those data type is 'object' will not plot in y-axis
    if col == 'date' or col == 'day' or Act4Dataset1[col].dtype == 'object':
        continue
    # Use plt.plot the data with Act4Dataset1['day'] in x-axis and Act4Dataset1[col] in y-axis,
    # and label each Act4Dataset1[col] for legend
    plt.plot(Act4Dataset1['day'], Act4Dataset1[col], label = col)

# Create a variable labelx and labely and store 'Day' and 'Number of Units Loss' in them.
# Purpose is to reuse the rest plot figure later
labelx = 'Day'
labely = 'Number of Units Loss'

# Label x-axis as labelx
plt.xlabel(labelx)
# Label y-axis as labely
plt.ylabel(labely)
# Label title as 'Daily Unit Lost'
plt.title('Daily Unit Loss')
# Activate the legend
plt.legend()
# Display the line plot
plt.show()
```

From the plot, APC is the most units lost while naval ship is the least.

```python
# We have already read the Dataset1 of Russia Ukraine war.
# Here you have to use import seaborn as sns library and sns.lineplot method.
# In addition to the Air-Vehicle, Russia has lost a large number of other weapons,
# including
# Tanks,
# Armoured Personnel Carriers (APCs),
# field Artillery,
# Multiple Rocket Launchers,
# military- automobiles,
# aircraft, and
# helicopters;

# visualize all of these weapons using a scatter plot 1 by 1 and all in one scatterplot
# that includes all of the possible parameters of Scatter plot .


# Create a list 'LargeWeapons' which is for loop purpose.
# Store required information 'tank', 'APC', 'field artillery', 'MRL', 'military auto', 'aircraft', 'helicopter' in it.
LargeWeapons = ['tank', 'APC', 'field artillery', 'MRL', 'military auto', 'aircraft', 'helicopter']

# Plot 'LargeWeapons' and display one by one plot figure
for Weapon in LargeWeapons:
    # set figure size with 10 width by 10 height in inches
    plt.figure(figsize = (10,10))
    # Use sns.scatterplot data with Act4Dataset1['day'] in x-axis and Act4Dataset1[Weapon] in y-axis
    sns.scatterplot(x = 'day', y = Weapon, data = Act4Dataset1)
    # Label x-axis as labelx
    plt.xlabel(labelx)
    # Label y-axis as labely
    plt.ylabel(labely)
    # Label title as 'Daily {Weapon} Unit Lost'
    plt.title(f'Daily {Weapon} Unit Loss')
    # Display the scatter plot
    plt.show()


# set figure size with 10 width by 10 height in inches
plt.figure(figsize = (10,10))

# Plot 'LargeWeapons' one by one into same one plot figure
for Weapon in LargeWeapons:
    # Use sns.scatterplot data with Act4Dataset1['day'] in x-axis and Act4Dataset1[Weapon] in y-axis,
    # and label each Act4Dataset1[Weapon] for legend
    sns.scatterplot(x = 'day', y = Weapon, data = Act4Dataset1, label = Weapon)


# Label x-axis as labelx
plt.xlabel(labelx)
# Label y-axis as labely
plt.ylabel(labely)
# Label title as 'Daily Weapon Unit Lost'
plt.title('Daily Weapon Unit Loss')
# Activate the legend
plt.legend()
# Display the scatter plot all in one
plt.show()
```

```python
# We have already read the Dataset1 of Russia Ukraine war.
# If you want you can read the file again using pandas
# To solve this tasks you have to use Sns.heatmap along with merge and corr method.
# Find out how each column is connected to the others since correlations are nothing more than determining how closely two variables are related
# Refer assignment to understand correlation. Since a correlation is nothing more than determining how closely two variables are related,
# find out how each column is related to the others by using corr method.


# Using Corr on Dataset1 and store them in 'Act4Dataset1Corr'
Act4Dataset1Corr = Act4Dataset1.corr(numeric_only=True)

#Display the content of 'Act4Dataset1Corr'
Act4Dataset1Corr
```

From the correlation data above, by observation, it is quite difficult identify which one is closer enough and which one not. Therefore, using heatmap method to easily observe the correlation.

```python
# You can plot heatmap to identify the correlation by using heatmap method.

# set figure size with 10 width by 10 height in inches
plt.figure(figsize = (10, 10))
# Use sns.heatmap for data Act4Dataset1Corr with annotation and the colourmap 'spring'
sns.heatmap(Act4Dataset1Corr, annot = True, cmap = 'spring')
# Label title as 'Heatmap for Dataset1'
plt.title('Heatmap for Dataset1')
# Display the heatmap
plt.show()
```

From above heatmap information,
 - 'mobile SRBM system' has the least relationship with others
 - 'tank' and 'vehicles and fuel tanks' have the most relationship with others.
 - There is no relationship between 'vehicles and fuel tanks', 'cruise missiles' and 'military auto', 'fuel tank', 'mobile SRBM system'

```python
# Since we have also read the other dataset, which is designated as Dataset2,
# Determine the daily death toll(the number of people who die in the war) by taking into account the date and the attribute personnel.

# Read "Dataset2.csv", store them in "Act4Dataset2"
Act4Dataset2 = pd.read_csv('Dataset2.csv')
# Using groupby to taking into account the date and the attribute personnel with sum
# if there is more than 1 data with same day.
Dataset2DatePersonnel = Act4Dataset2.groupby('date')['personnel'].sum()

#Display the content 'Dataset2DatePersonnel'
Dataset2DatePersonnel
```

```python
# You can plot the histogram to analyze the daily death toll for the given task.

# Plot histogram with Dataset2DatePersonnel and set bins to 10
plt.hist(Dataset2DatePersonnel, bins = 10)
# Label x-axis as 'Daily death toll'
plt.xlabel('Daily death toll')
# Label y-axis as 'Frequency'
plt.ylabel('Frequency')
# Label x-axis as 'Analyze the daily death toll'
plt.title('Analyze the daily death toll')
# Activate the guideline
plt.grid()
# Display the histogram
plt.show()
```

From histogram above, it has more than 35 times between 10k to 50k on daily death toll.
